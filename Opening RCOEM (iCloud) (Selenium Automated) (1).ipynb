```python
threshold = 60
for i in range(len(subjects)):
    a, total = map(int, attended[i].split('/'))
    
    if (a / total) * 100 < threshold:
        # Find how many more classes to attend
        extra_classes = 0
        while (a / total) * 100 < threshold:
            a += 1
            total += 1
            extra_classes += 1
        print(f"{subjects[i]}: Attend at least {extra_classes} more classes to cross 60%")
    
    else:
        # Find how many classes can be bunked
        bunkable = 0
        temp_a, temp_total = a, total
        while ((temp_a / temp_total) * 100) >= threshold:
            temp_a -= 1
            temp_total -= 1
            bunkable += 1
        
        # The last decrement takes it below 60%, so we subtract one from the bunkable count
        bunkable -= 1  
        print(f"{subjects[i]}: You can bunk {bunkable} classes and still stay above 60%")
    print()
```
